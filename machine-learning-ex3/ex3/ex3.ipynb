```python
%matplotlib notebook
import numpy as np
import matplotlib.pylab as plt
import scipy.io as sio

data = sio.loadmat("ex3data1.mat")
X, y = data['X'], data['y']
row, col = 10, 10
X_some = np.random.permutation(data['X'])[:row*col]
X_some = X_some.reshape((row,col,20,20), order='F')

def img_row(X):
    for x in X:
        yield np.concatenate(x, axis=1)
        
imgdata = np.concatenate(list(img_row(X_some)))
img = plt.imshow(imgdata, cmap='gray')
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(C=10)
clf.fit(X, y.ravel())

print("The training set accuracy is {:.2%}".format(clf.score(X,y)))
```
