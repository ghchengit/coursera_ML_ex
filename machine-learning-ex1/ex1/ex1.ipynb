```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt


def plot_linear_regression_1d(x, y, x1=None, y1=None, title='', xlabel='', 
                              ylabel='', style=('xr', '-b')):
    fig, ax = plt.subplots()
    ax.plot(x, y, style[0], label="training data")
    if x1 is not None and y1 is not None:
        ax.plot(x1, y1, style[1], label="linear regression")
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_legend()
    
    
data = np.genfromtxt("ex1data1.txt", delimiter=",")
x, y = data[:, 0], data[:, 1]
plot_linear_regression_1d(
    x=x,
    y=y,
    xlabel="Population of City in 10,000s",
    ylabel="Profit in $10,000s",
)
```

```python
from sklearn import linear_model

#x_stack = np.hstack((x.reshape(-1, 1), np.ones(len(x)).reshape(-1,1)))
regr = linear_model.LinearRegression()
regr.fit(x.reshape(-1, 1), y)
x1 = np.linspace(min(x), max(x), 100)
y1 = regr.predict(x1.reshape(-1, 1))

plot_linear_regression_1d(
    x=x,
    y=y,
    x1=x1,
    y1=y1,
    xlabel="Population of City in 10,000s",
    ylabel="Profit in $10,000s",
)
```
